# 基于pandas的数据可视化

##  pandas绘图介绍

### pandas绘图简介

>在Python中有很多图形化库，例如之前讲述的matplotlib库。  
虽然matplotlib库可以绘制精美的图形，但需要安装大量的组件，书写大量的代码，并且绘图过程也比较复杂。  
在pandas中则可以高效的完成绘图的工作，因此本节将讲述使用pandas来绘制图形。

### pandas绘图原理

>pandas使用一维的数据结构Series和二维的数据结构DataFrame来表示数据，因此与numpy相比，pandas可以存储混合的数据结构。  
同时pandas使用NaN来表示缺失的数据，因而制作一张完整的图表，只需几条语句即可实现。

**pandas库导入语句：**

>import pandas as pd  
   import numpy as np  
   from pandas import DataFrame,Series  
   import matplotlib.pyplot as plt  

**pandas库中的基本数据类型及含义**

|数据类型|含义|
|---|---|
|Series| pandas库中的一维数组|
|DataFrame| pandas库中的二维数组|

**专用于series的plot方法参数**

|参数|含义|
|---|---|
|label| 图表的标签|
|ax |要进行绘制的matplotlib subplot对象|
|style| 要传给matplotlib的字符风格|
|alpha |图表的填充透明度（0~1）|
|kind |图表的类型，有line，bar，barh及kde等|
|logy| 在y轴上使用对数标尺|
|rot |旋转刻度标签（0~360°）|
|xticks、yticks| 用作x轴和y轴刻度的值|
|xlim、ylim |x轴和y轴的界限|

```python
import pandas as pd
import numpy as np

# 生成示例数据
data = np.random.randn(100)
s = pd.Series(data)

# 绘制Series
s.plot(label='Random Data',  # 图表的标签
       ax=None,  # 要进行绘制的matplotlib subplot对象,默认为None将创建新图
       style='-',  # 要传给matplotlib的字符风格,如'-'线形,':'虚线等
       alpha=0.8,  # 图表的填充透明度(0~1)
       kind='line',  # 图表的类型,有line,bar,barh及kde等
       logy=False,  # 在y轴上使用对数标尺
       use_index=True,  # 使用Series的索引作为x轴
       rot=0,  # 旋转刻度标签(0~360°)
       xticks=None,  # 用作x轴刻度的值
       yticks=None,  # 用作y轴刻度的值
       xlim=None,  # x轴的界限
       ylim=None,  # y轴的界限
       grid=False  # 是否显示网格线
       )

# 显示图形
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 生成示例数据
data = np.random.randn(100)
s = pd.Series(data)

# 绘制Series
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 默认设置
s.plot(ax=axes[0][0], title='Default Settings')

# 自定义刻度和限制
xticks = np.arange(0, 101, 20)
yticks = np.linspace(-3, 3, 7)
xlim = (10, 90)
ylim = (-2.5, 2.5)
s.plot(ax=axes[0][1], xticks=xticks, yticks=yticks, xlim=xlim, ylim=ylim, title='Custom Ticks and Limits')

# 对数刻度和网格线
s.plot(ax=axes[1][0], logy=True, grid=True, title='Log Scale and Grid')

# 自定义样式和透明度
s.plot(ax=axes[1][1], style='--', alpha=0.5, title='Custom Style and Alpha')

# 调整布局
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 设置默认绘图样式
sns.set(style="darkgrid", palette="muted", context="notebook", font_scale=2)


# 绘制散点图
sns.relplot(x="total_bill", y="tip", data=tips)

plt.show()
```

```python
# 创建子图
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成示例数据
data = np.random.normal(loc=0, scale=1, size=100)

# 设置 Seaborn 风格
sns.set_style("dark")#'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'

# 绘制核密度估计图
sns.kdeplot(data, fill=True)

# 自定义图形样式
plt.title("Kernel Density Estimation", fontsize=16)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Density", fontsize=14)

# 展示图形
plt.show()
```

- 使用 sns.axes_style() 的好处是可以在代码块内部临时设置特定的样式，而不会影响到其他部分的绘图样式。  
- 对于在同一个脚本或notebook中绘制具有不同样式的多个图像非常有用。
- 需要注意的是，sns.axes_style() 只会影响图像的轴样式，而不会影响其他绘图元素的样式，如调色板、字体大小等。  
- 如果需要同时设置这些其他样式，可以结合使用 sns.set_style()、sns.set_palette() 等函数。

```python
import seaborn as sns
import matplotlib.pyplot as plt


# 加载数据集
tips = sns.load_dataset("tips",data_home='Data',cache=True)

# 使用默认样式绘制第一个图像
plt.figure()
sns.scatterplot(x="total_bill", y="tip", data=tips)

# 使用 axes_style 上下文管理器设置样式并绘制第二个图像
with sns.axes_style("ticks"):
    plt.figure()
    sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.show()
```

**Seaborn 提供了四种预设的上下文环境,分别适用于不同的使用场景:**

- "paper": 适用于最终版本的出版物,如学术论文或报告。字体相对较小,以便节省空间。
- "notebook": 适用于交互式编码环境,如 Jupyter Notebook。字体稍大于 "paper" 环境。
- "talk": 适用于演示文稿或报告。字体比 "notebook" 环境更大,以便在远处观看时清晰可见。
- "poster": 适用于海报展示。字体最大,以便在远处观看时清晰可见。

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载数据集
tips = sns.load_dataset("tips",data_home='Data',cache=True)

# 设置 "talk" 上下文环境,字体放大 1.2 倍
sns.set_context("talk", font_scale=1.2)

# 绘制散点图
sns.relplot(x="total_bill", y="tip", data=tips)

plt.show()
```

- 与 color_palette("hls") 相比,hls_palette 函数提供了更多控制选项,包括亮度、饱和度、色调范围和颜色数量。  
- 这使得它在生成具有特定颜色属性的调色板时更加灵活。  
- 需要注意的是,与其他一些调色板不同,hls_palette 生成的颜色可能会存在重叠,因此在使用时需要根据具体情况评估是否合适。  
- 但是,如果需要生成具有一致亮度和饱和度的颜色集合,hls_palette 就是一个非常好的选择。  

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 生成默认的 HLS 调色板
hls_pal = sns.hls_palette()
print(hls_pal)

# 绘制调色板
sns.palplot(hls_pal)
plt.show()
```

```python
# 生成 10 种颜色的 HLS 调色板
hls_pal = sns.hls_palette(n_colors=10)
sns.palplot(hls_pal)
plt.show()
```

```python
# 生成亮度较低、饱和度较高的 HLS 调色板
hls_pal = sns.hls_palette(l=0.4, s=0.8)
sns.palplot(hls_pal)
plt.show()
```

**在 hls_palette 函数中, h 参数用于指定生成调色板时所使用的色调(Hue)范围。色调范围是一个介于 0 到 1 之间的浮点数。**

**具体来说:**

- h=0 表示只使用红色色调
- h=1 表示使用整个色环,包括红、橙、黄、绿、蓝、紫等所有色调

### 脊图

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Create the data
rs = np.random.RandomState(1979)
x = rs.randn(500)
g = np.tile(list("ABCDEFGHIJ"), 50)
df = pd.DataFrame(dict(x=x, g=g))
m = df.g.map(ord)
df["x"] += m

# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(df, row="g", hue="g", aspect=15, height=1.5, palette=pal)  # 增加 height 为 1.5

# Draw the densities in a few steps
g.map(sns.kdeplot, "x",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "x", clip_on=False, color="w", lw=2, bw_adjust=.5)

# passing color=None to refline() uses the hue mapping
g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)

# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "x")

# Set the subplots to overlap
g.figure.subplots_adjust(hspace=0.5)  # 增加 hspace 为 0.5

# Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)

plt.show()
```

### 对角线相关矩阵

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# Generate a large random dataset
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 columns=list(ascii_letters[26:]))

# Compute the correlation matrix
corr = d.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
